```python
df_spot = pd.read_csv('/kaggle/input/spotify-data-from-pyspark-course/spotify-data.csv')
df_spot
```

## The most popular songs in the last 10 years

```python
current_year = pd.Timestamp.now().year
```

```python
df_spot['year'] = pd.to_datetime(df_spot['year'], format='%Y')
```

```python
last_10_years_df = df_spot[df_spot['year'].dt.year.between(current_year - 10, current_year)]
sorted_df = last_10_years_df.sort_values(by='popularity')
```

```python
subset_df = sorted_df[['name', 'artists']].head(10)
subset_df
   
```

## The number of songs released by artist and year of release

```python
id_counts = df_spot.groupby(['artists', 'year'])['id'].count().reset_index()
```

```python
id_counts = id_counts.rename(columns={'id': 'numb of songs'})
id_counts = id_counts.sort_values(by='numb of songs', ascending=False)
id_counts
```

## The artists with the highest average song popularity over the last 5 years.

```python
last_5_years_df = df_spot[df_spot['year'].dt.year.between(current_year - 5, current_year)]
```

```python
popularity_count = last_5_years_df.groupby(['artists'])['popularity'].sum().reset_index()
```

```python
popularity_count = popularity_count.sort_values(by='popularity', ascending=False)
popularity_count
```
